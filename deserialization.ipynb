# Deserialization

```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    first_name: str
    last_name: str
    age: int
```

```python
data = {
    "first_name": "John",
    "last_name": "Doe",
    "age": 30
}
p = Person.model_validate(data)

# OR
missing_data = {
    "first_name": "John",
    "last_name": "Doe"
}
try:
    p = Person.model_validate(missing_data)
except Exception as e:
    print(e)
```
