```python
# instead of using the default values use optional or default types because they are mutable and can cause unexpected behavior.
class Person(BaseModel):
    name: str
    age: int
    hobbies: List[str] = []  # This is a mutable default value, which can lead to unexpected behavior.
```

```python
class Person(BaseModel):
    name: str
    age: int
    hobbies: Optional[List[str]] = None  # This is an optional field with a default value of None.
```
