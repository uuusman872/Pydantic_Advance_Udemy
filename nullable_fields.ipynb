```python
from pydantic import BaseModel, ValidationError
from time import time
from typing import List, Optional, Union
```

```python
# Nullable fields can be defined using the type hinting syntax with None
class Model(BaseModel):
    field: int | None

# Another way for nullable fields is to use Optional from typing
class ModelOptional(BaseModel):
    field: Optional[int] = None

# Another way for nullable fields is to use default value None
class ModelDefault(BaseModel):
    field: int | None = None

# with union
class ModelUnion(BaseModel):
    field: Union[int, None] = None
```

```python
try:
    Model(field=None)
except ValidationError as e:
    print(e)
```
