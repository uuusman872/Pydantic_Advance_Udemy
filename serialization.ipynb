# Serialization

```python
from pydantic import BaseModel, ValidationError

class Person(BaseModel):
    first_name: str
    last_name: str
    age: int
```

```python
galois = Person(first_name="Glois", last_name="Smith", age=20)
newton = Person(first_name="Isaac", last_name="Newton", age=84)
```

```python
# do not use the __dict__ method to access the attributes of the model, use the model_dump() method instead
newton.__dict__
```

```python
print(galois.model_dump())
print(type(galois.model_dump()))
print(galois.model_dump_json(indent=2))
print(type(galois.model_dump_json(indent=2)))
print(galois.model_dump(exclude=["age"]))
print(galois.model_dump_json(include=["age"], indent=2))
```
