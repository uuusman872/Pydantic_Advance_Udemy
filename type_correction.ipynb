```python
from pydantic import BaseModel, ValidationError

class Coordinates(BaseModel):
    x : float
    y : float
    b : bool

```

```python
p1 = Coordinates(x="1", y="2", b="1")
p1
```

```python
# Example why we should use strict types in pydantic models to avoid unexpected type coercion and potential bugs in our code.
# Strings get converted to integers, floats but intergers can not be converted to strings, floats or booleans. This can lead to unexpected behavior and make it harder to debug issues in our code.

class Model(BaseModel):
    field: str

try:
    m = Model(field=123)
except ValidationError as e:
    print(e)
```

```python
class Contact(BaseModel):
    email: str

new_json_data = """
{
    "email": {
        "personal": "john.doe@example.com",
        "work": "john.doe@company.com"
    }
}
"""

try:
    contact = Contact.model_validate_json(new_json_data)
except ValidationError as e:
    print(e)
```

```python
new_data_dict = {
    "email": {
        "personal": "john.doe@example.com",
        "work": "john.doe@company.com"
    }
}

try:
    contact = Contact(email=str(new_data_dict["email"]))
    print(contact)
except ValidationError as e:
    print(e)
```
